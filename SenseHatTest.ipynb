```python
# import packages 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Optional - Set figure resolution to 300dpi
import matplotlib as mpl
%matplotlib inline
plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300
```

## Just testing out the Sense Hat by picking up the Raspberry pi and moving it around for a few seconds

```python
df = pd.read_csv (r'./data.csv', index_col=13) 
df
```

#### A look at the total data collected:

```python
df.plot.line(marker = 'o')
```

#### Now to looks at some specifics:

```python
fig, ax = plt.subplots()
df.plot.line(ax = ax, y = 'pitch',color = 'r', marker = 'o')
df.plot.line(ax = ax, y = 'roll',color = 'b', marker = 'o')
df.plot.line(ax = ax, y = 'yaw',color = 'y', marker = 'o')
ax.set_xlabel("Time (msec)")
ax.set_ylabel("Angle ($^\circ$)")
plt.title('Orientation over time')
```

```python
fig, ax = plt.subplots()
df.plot.line(ax = ax, y = 'acc_x',color = 'r', marker = 'o')
df.plot.line(ax = ax, y = 'acc_y',color = 'b', marker = 'o')
df.plot.line(ax = ax, y = 'acc_z',color = 'y', marker = 'o')
ax.set_xlabel("Time (msec)")
ax.set_ylabel("Acceleration (g)")
plt.title('Acceleration over time')
```

```python
fig, ax = plt.subplots()
df.plot.line(ax = ax, y = 'hum',color = 'r', marker = 'o')
ax.set_xlabel("Time (msec)")
ax.set_ylabel("%")
plt.title('Humidity over time')
```

```python
fig, ax = plt.subplots()
df.plot.line(ax = ax, y = 'temp',color = 'r', marker = 'o')
ax.set_xlabel("Time (msec)")
ax.set_ylabel("$^\circ$C")
plt.title('Temperature over time')
```

```python
fig, ax = plt.subplots()
df.plot.line(ax = ax, y = 'pres',color = 'r', marker = 'o')
ax.set_xlabel("Time (msec)")
ax.set_ylabel("millibars")
plt.ticklabel_format(useOffset=False)
plt.title('Pressure over time')
```
